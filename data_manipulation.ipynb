## Remote job in 2023

```python
tableRemote = data[['experience_level', 'job_title', 'remote_ratio', 'company_size']]
tableRemote = tableRemote[tableRemote['experience_level'] == 'EN']
tableRemote
```

Large Companies

```python
tableRemoteL = tableRemote[tableRemote['company_size'] == 'L']
tableRemoteL.groupby('job_title')['remote_ratio'].mean()
```

This a percentage table of remotly employment for large companies

Medium

```python
tableRemoteM = tableRemote[tableRemote['company_size'] == 'M']
tableRemoteM.groupby('job_title')['remote_ratio'].mean()
```

Small

```python
tableRemoteS = tableRemote[tableRemote['company_size'] == 'S']
percentageSCompanies = tableRemoteS.groupby('job_title')['remote_ratio'].mean()
percentageSCompanies
```

```python
percentageYear = data.groupby('work_year')['remote_ratio'].mean()
percentageYear
```

Remote jobs in data science seem to decrease by year. It had a 65% of employability in 2020 maybe due to COVID19 Pandemic.
